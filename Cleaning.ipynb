```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
import numpy as np
```

```python
data = pd.read_csv('data_offer.csv')
```

```python
df = data.copy()
```

```python
df = df.dropna()
df
```

```python
df["salary"] = df["salary"].str.replace("\n", "")
df["salary"] = df["salary"].str.replace(",", "")
df["salary"] = df["salary"].str.replace("€", "")
```

```python
df
```

```python
year_salaries = df[df["salary"].str.contains("an")]
month_salaries = df[df["salary"].str.contains("mois")]
day_salaries = df[df["salary"].str.contains("jour")]
hour_salaries = df[df["salary"].str.contains("heure")]
```

```python
year_salaries["salary"] = year_salaries["salary"].str.replace(" par an", "")
month_salaries["salary"] = month_salaries["salary"].str.replace(" par mois", "")
day_salaries["salary"] = day_salaries["salary"].str.replace(" par jour", "")
hour_salaries["salary"] = hour_salaries["salary"].str.replace(" par heure", "")
```

```python
def min_sal(sal):
    try:
        splt = sal.split(' - ', 1)
        first = float(splt[0])
        return first
    except:
        return float(sal)

def max_sal(sal):
    try:
        splt = sal.split(' - ', 1)
        second = float(splt[1])
        return second
    except:
        return float(sal)
    
def avg_sal(sal):
    try:
        splt = sal.split(' - ', 1)
        first = float(splt[0])
        second = float(splt[1])
        return (first + second)/2
    except:
        return float(sal)
```

```python
year_salaries['salary'] = year_salaries['salary'].str.replace('\xa0','')
```

```python
year_salaries["salary"] = year_salaries["salary"].apply(avg_sal)
```

```python
month_salaries["salary"] = month_salaries["salary"].str.replace('\xa0','')
day_salaries["salary"] = day_salaries["salary"].str.replace('\xa0','')
hour_salaries["salary"] = hour_salaries["salary"].str.replace('\xa0','')
```

```python
month_salaries["salary"] = month_salaries["salary"].apply(avg_sal)
day_salaries["salary"] = day_salaries["salary"].apply(avg_sal)
hour_salaries["salary"] = hour_salaries["salary"].apply(avg_sal)
```
